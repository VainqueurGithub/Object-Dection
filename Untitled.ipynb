```python
def predict_class(filename):
    model = YOLO('runs/detect/yolov8m_custom8/weights/best.pt')

    #Predicting
    result=model.predict(
        source="C:/Users/vainq/Wild animal detection/inference_data/"+filename,show = True, imgsz=640,save = True,name='yolov8m')
    return result
```
